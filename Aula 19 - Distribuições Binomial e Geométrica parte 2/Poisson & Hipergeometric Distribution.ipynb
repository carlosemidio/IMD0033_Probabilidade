# Distribuição de Poisson

## pmf

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import linspace
from scipy.stats import poisson

# load the seaborn configurations
sns.set()

# Create a range of numbers from 0 to 30, with 31 elements (each number has one entry).
outcome_counts = linspace(0,10,11)

# Create the pmf, one for each entry in outcome_counts.
outcome_probs1 = poisson.pmf(outcome_counts,1)
outcome_probs2 = poisson.pmf(outcome_counts,4)
outcome_probs3 = poisson.pmf(outcome_counts,10)

fig = plt.figure(figsize=(20,12))

ax1 = fig.add_subplot(3,3,1)
ax2 = fig.add_subplot(3,3,2)
ax3 = fig.add_subplot(3,3,3)

ax1.set_title("Lambda = 1");
ax2.set_title("Lambda = 4");
ax3.set_title("Lambda = 10")

ax1.bar(outcome_counts, outcome_probs1)
ax2.bar(outcome_counts, outcome_probs2)
ax3.bar(outcome_counts, outcome_probs3)

plt.show()
```

## cdf

```python
import matplotlib.pyplot as plt
from scipy import linspace
from scipy.stats import poisson
# Create a range of numbers from 0 to 30, with 31 elements (each number has one entry).
outcome_counts = linspace(0,10,11)

# Create the pmf, one for each entry in outcome_counts.
outcome_probs1 = poisson.cdf(outcome_counts,1)
outcome_probs2 = poisson.cdf(outcome_counts,4)
outcome_probs3 = poisson.cdf(outcome_counts,10)

fig = plt.figure(figsize=(20,12))

ax1 = fig.add_subplot(3,3,1)
ax2 = fig.add_subplot(3,3,2)
ax3 = fig.add_subplot(3,3,3)

ax1.set_title("Lambda = 1");
ax2.set_title("Lambda = 4");
ax3.set_title("Lambda = 10")

ax1.bar(outcome_counts, outcome_probs1)
ax2.bar(outcome_counts, outcome_probs2)
ax3.bar(outcome_counts, outcome_probs3)

plt.show()
```

# Distribuição Poisson - Aproximação

```python
from scipy import linspace
from scipy.stats import binom
from scipy.stats import poisson

# Create a range of numbers from 0 to 30, with 31 elements (each number has one entry).
outcome_counts = linspace(0,20,21)

# Create the pmf, one for each entry in outcome_counts.
outcome_probs = binom.pmf(outcome_counts,20,0.05)
outcome_probs1 = poisson.pmf(outcome_counts,1)

red_line = plt.scatter(outcome_counts, outcome_probs, color='r',label='X~binomial(20,0.05)')
blue_line = plt.scatter(outcome_counts, outcome_probs1, color='b',label='Y~poisson(20*0.05)')

plt.legend(handles=[red_line, blue_line])

plt.show()
```

# Distribuição Hipergeométrica

## Distribuição Hipergeométrica - pmf

```python
import matplotlib.pyplot as plt
from scipy import linspace
from scipy.stats import hypergeom
import numpy as np


[M, n, N] = [100,5,4]
rv = hypergeom(M, n, N)
x = np.arange(0, 21)
pmf_dogs = rv.pmf(x)
fig = plt.figure()
blue_line = plt.plot(x, pmf_dogs, 'bo', color='b',label='Y~hypergeom(k,100,5,4)')
plt.legend(handles=blue_line)

plt.show()
```
